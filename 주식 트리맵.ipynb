```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

import pandas as pd
import numpy as np
import FinanceDataReader as fdr
from pykrx import stock
```

```python
# 날짜는 바꿔줘야 합니다~~~~~ 
```

```python
# 코스피 등락률 가져오기 
df=stock.get_market_price_change_by_ticker(fromdate="20221011", todate="20221011")
df['종목코드']=df.index
```

```python
# 코스피 섹터 가져오기 
stocks = fdr.StockListing('KOSPI')
sector = stocks[['Symbol','Sector']]
sector = sector.to_numpy()
```

```python
# 코스피 사이즈 가져오기 
size = stock.get_market_cap('20221011')
# ticker를 안에 넣어주기 
size['Symbol']= size.index
size = size[['Symbol','시가총액']]
size = size.to_numpy()
```

```python
# 등락률과 섹터 합치는 함수 
def make_sector(code):
    for i in range(len(sector)):
        if code == sector[i][0]:
            return sector[i][1]
    return 'No'
```

```python
# 등락률과 사이즈 합치는 함수 
def make_size(code):
    for i in range(len(size)):
        if code == size[i][0]:
            return size[i][1]
    return 'No'
```

```python
#합치기 
df['섹터'] = df['종목코드'].map(make_sector)
df['시가총액'] = df['종목코드'].map(make_size)
```

```python
# 필요없는 columns 버리기 
df=df.drop(['시가','종가','변동폭','거래량','거래대금'],axis='columns')
```

```python
#코스피인거 나타내주기 위해 'kospi' 추가해주기
df['시장구분']='KOSPI'
```

```python
# 섹터 nan을 '우선주' 로 채워주기 
df.loc[df['섹터'].isna(),'섹터']='우선주'
```

```python
# Treemap 그리기
fig = px.treemap(df, path=['시장구분','섹터','종목명'], values='시가총액',
                color='등락률',
                 hover_data=['등락률','종목코드'],
                 
                 range_color=(-0.3,0.3),color_continuous_scale='RdBu')
fig.update_traces(textposition='middle center',text = df['등락률'])
fig.update_traces(textfont_color='white',textfont_size=25)

fig.update_layout(height=800)
fig.show()
```
