# 뉴스 크롤링

```python
#웹페이지 스크립트 가져오기
date = input('날짜를 입력해주세요(ex.20221021):')  #날짜에 따라 크롤링 가능
url = "https://finance.naver.com/news/mainnews.naver?"  
params = {'date' : date}

resp = requests.get(url, params)
soup = BeautifulSoup(resp.content, 'lxml')

#각 뉴스기사의 링크 크롤링하기
url_list = soup.find('div','mainNewsList').find_all('dl')
# url_list = soup.find('div','mainNewsList').find_all('a')
news_url_list = []

for url in url_list :
    news_url_list.append('https://finance.naver.com' + url.find('dt').find('a')['href'])
```

```python
# pip install newspaper3k
from newspaper import Article
import re

# 뉴스 텍스트 크롤링 및 문장 단위 분리
def article_crawling(url):
    resp = requests.get(url)
    soup = BeautifulSoup(resp.content, 'lxml')
    article = soup.find('div','articleCont')
    while article.b != None:
        article.find('b').decompose()
    while article.span != None:
        article.find('span').decompose()
    while article.div != None:
        article.find('div').decompose()
    while article.strong != None:
        article.find('strong').decompose()         
    try:
        article.find('div','link_news').decompose()
    except:
        pass
    article = article.text
    article1 = re.split('(다\.)',article)
    article2 = []
    for i in range(len(article1)):
        try:
            if (i%2) == 0:
                article2.append((article1[i] + article1[i+1]).lstrip().rstrip())
        except:
            pass
        
    return article2

# 뉴스 기사 텍스트 크롤링 반복 함수
def articles(news_url_list):
    articles = []
    for i in news_url_list:
        articles.append(article_crawling(i))
    return articles

# 제목 크롤링
def title_crawling(news_url_list):
    title = []
    for i in news_url_list:
        resp = requests.get(i)
        soup = BeautifulSoup(resp.content, 'lxml')
        t = soup.find('div','article_info').find('h3').text.lstrip().rstrip()
        if t[0]=="[":
            t = "".join(t.split(']')[1:]).lstrip().rstrip()
        elif t[-1]=="]":
            t = "".join(t.split('[')[:-1]).lstrip().rstrip()
        t = t.lstrip('"').rstrip('"')
        title.append(t)
        
    return title

# 기자 크롤링
def reporter_crawling(urls):
    reporter = []
    for url in urls:
        resp = requests.get(url)
        soup = BeautifulSoup(resp.content, 'lxml')
        try:
            reporter.append(soup.find('div','info_thumb').find('span','name').text)
        except:
            reporter.append(soup.find('div','card_journalist_byline').text.split(' ')[0].lstrip())
    return reporter

# 뉴스사 크롤링
def source_crawling(urls):
    source = []
    for url in urls:
        resp = requests.get(url)
        soup = BeautifulSoup(resp.content, 'lxml')
        source.append(soup.find('div','sponsor').find('img')['title'])
    return source

import pandas as pd
df = pd.DataFrame()
df['news_url'] = news_url_list
df['news_date'] = date[:4]+'-'+date[4:6]+'-'+date[6:]
df['news_title'] = title_crawling(news_url_list)
df['news_sum1'] = 0
df['news_sum2'] = 0
df['news_sum3'] = 0
df['news_src'] = source_crawling(news_url_list)
df['news_rep'] = reporter_crawling(news_url_list)
```

```python
import re
# 불용어제거
stopwords = ['/뉴스1' , "ⓒ", "News1","/AP=뉴시스","/연합뉴스","연합뉴스",'[파이낸셜뉴스]','━━']
dropwords = ['/AFPBBNews=뉴스1',"#기자페이지","어썸머니","3분 월스트리트",'◆']

# 불용어제거
def s_repeat(news_article):
    for i in range(len(news_article)):
        if len(news_article[i].split(" = ")) >1:
            news_article[i] = news_article[i].split(" = ")[1]
        if len(news_article[i].split("기자]")) >1:
            news_article[i] = news_article[i].split("기자]")[1]
        for dw in dropwords:
            if dw in news_article[i]:
                news_article[i] = ''
        for sw in stopwords:
            news_article[i] = news_article[i].replace(sw,'')
            
        news_article[i] = news_article[i].lstrip().rstrip()
    
    news_article = list(filter(None, news_article))
    news_article = list(dict.fromkeys(news_article))
    return news_article

# 길이에 따라 문장 합침
def text2sentences(news_article):
    for idx in range(len(news_article)-1):
        if len(news_article[idx]) <= 30:
            news_article[idx+1] = (news_article[idx]+' '+news_article[idx+1])
            news_article[idx] = ''
    if len(news_article[-1]) <= 30:
            news_article[-2] += (' '+news_article[-1])
            news_article[-1] = ''
    news_article = list(filter(None, news_article))
    news_article = list(dict.fromkeys(news_article))
    return news_article

for i in range(len(news_articles)):
    news_articles[i] = s_repeat(news_articles[i])
    
for i in range(len(news_articles)):
    news_articles[i] = text2sentences(news_articles[i])
```

```python
tfidf = TfidfVectorizer()
cnt_vec = CountVectorizer()
graph_sentence = []

def build_sent_graph(sentence):
    tfidf_mat = tfidf.fit_transform(sentence).toarray()
    graph_sentence = np.dot(tfidf_mat, tfidf_mat.T)
    return graph_sentence

def build_words_graph(sentence):
    cnt_vec_mat = normalize(cnt_vec.fit_transform(sentence).toarray().astype(float), axis=0)
    vocab = cnt_vec.vocabulary_
    return np.dot(cnt_vec_mat.T, cnt_vec_mat), {vocab[word] : word for word in vocab}
    
def get_ranks(graph, d=0.85): # d = damping factor
    A = graph
    matrix_size = A.shape[0]
    for id in range(matrix_size):
        A[id, id] = 0 # diagonal 부분을 0으로
        link_sum = np.sum(A[:,id]) # A[:, id] = A[:][id]
        if link_sum != 0:
            A[:, id] /= link_sum
        A[:, id] *= -d
        A[id, id] = 1

    B = (1-d) * np.ones((matrix_size, 1))
    ranks = np.linalg.solve(A, B) # 연립방정식 Ax = b
    return {idx: r[0] for idx, r in enumerate(ranks)}

def summarize(news_article,sent_num=3):
    summary = []
    index=[]
    for idx in sorted_sent_rank_idx[:sent_num]:
        index.append(idx)
    index.sort()
    for idx in index:
        summary.append(news_article[idx])
    return summary
```
