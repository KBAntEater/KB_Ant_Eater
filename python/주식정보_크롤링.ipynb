# 코스피 종목 크롤링

```python
import time
import pandas as pd
stock_code = stock.get_market_ticker_list() # 현재일자 기준 가장 가까운 영업일의 코스피 상장종목 리스트
res = pd.DataFrame()
cnt = 0 # 크롤링 주식 종목 확인 용
for ticker in stock_code:
    cnt+=1
    print(cnt)
    print(stock.get_market_ticker_name(ticker))
    df = stock.get_market_ohlcv_by_date(fromdate="20000101", todate="20221020", ticker=ticker)
    df = df.assign(종목코드=ticker, 종목명=stock.get_market_ticker_name(ticker))
    res = pd.concat([res, df], axis=0)
    time.sleep(1)
res = res.reset_index()
res = res[['종목코드','종목명','날짜','시가','고가','저가','종가','거래량']]
res.columns = ['s_ticker','s_name','s_date','open','high','low','close','s_volume']
```

```python
def add_0(x):
    x = str(x)
    if len(x) == 5:
        x = '0'+x
    if len(x) == 4:
        x = '00'+x
    if len(x) == 3:
        x = '000'+x
    if len(x) == 2:
        x = '0000'+x
    if len(x) == 1:
        x = '00000'+x
    return x

res['s_ticker'] = res['s_ticker'].apply(lambda x: add_0(x))
```

```python
res.to_csv('stock_db.csv', index=False)
```
