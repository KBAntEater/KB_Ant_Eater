```python
from selenium import webdriver
driver = webdriver.Chrome('chromedriver.exe')
driver.implicitly_wait(3)
```

```python
import pandas as pd
import time

df = pd.DataFrame(columns=['테마명','종목명'])

for i in range(1,8):
    addr = f'https://finance.naver.com/sise/theme.naver?&page={i}'
    driver.get(addr)
    theme_list = driver.find_elements_by_xpath('//*[@id="contentarea_left"]/table[1]/tbody/tr')
    for j in range(3,len(theme_list)):
        if len(driver.find_elements_by_xpath(f'//*[@id="contentarea_left"]/table[1]/tbody/tr[{j}]/td')) == 8:
            tn = driver.find_element_by_xpath(f'//*[@id="contentarea_left"]/table[1]/tbody/tr[{j}]/td/a').text
            driver.find_element_by_xpath(f'//*[@id="contentarea_left"]/table[1]/tbody/tr[{j}]/td/a').click()
            stock_list = driver.find_elements_by_xpath('//*[@id="contentarea"]/div[4]/table/tbody/tr')
            temp = []
            for k in range(len(stock_list)):
                if len(driver.find_elements_by_xpath(f'//*[@id="contentarea"]/div[4]/table/tbody/tr[{k}]/td')) == 11:
                    temp.append(driver.find_element_by_xpath(f'//*[@id="contentarea"]/div[4]/table/tbody/tr[{k}]/td/div').text)
            temp_df = pd.DataFrame()
            temp_df['종목명'] = temp
            temp_df['테마명'] = tn
            
            df = pd.concat([df,temp_df])
            print(tn)
            driver.get(addr)
```

```python
kospi_df = df[~df['종목명'].str.contains('\*')]
```

```python
kospi_df.drop_duplicates().to_csv('stock_theme_kospi.csv',index=False)
```
